## Exercise 1: Exceptional Olympians

Scrape data from [this wikipedia site](https://en.wikipedia.org/wiki/List_of_multiple_Olympic_medalists) about exceptional Olympic medalists. 

1. Download the html using urllib. 
2. Parse this html with BeautifulSoup.
3. Extract the html that corresponds to the big table from the soup.
4. Parse the table into a pandas dataframe. Hint: both the "No." and the "Total." column use row-spans which are tricky to parse, both with a pandas reader and manually. For the purpose of this exercise, exclude all rows that are not easy to parse (the first one is Bjørn Dæhlie).
5. Create a table that shows for each country how many gold, silver, bronze, and total medals it won in that list.

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

Download the html using urllib. 
Parse this html with BeautifulSoup.

```python
# 1 Download the html using urllib
url = "https://en.wikipedia.org/wiki/List_of_multiple_Olympic_medalists"

with urllib.request.urlopen(url) as response:
    html = response.read()
    html = html.decode('utf-8')

# 2 Parse this html with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# 3 Extract the html that corresponds to the big table from the soup
data = soup.select("table.wikitable")[0]
```

```python
# 4 Parse the table into a pandas dataframe
headers_base = data.find_all('th')
headers = pd.Series([tag.get_text() for tag in headers_base]).str.replace('\n','')
rows_base = data.select('td')
rows = [tag.get_text() for tag in rows_base]

# Set up lists for adding data
count = 0
no = []
athlete = []
nation = []
sport = []
years = []
games = []
gender = []
gold = []
silver = []
bronze = []

gender = ['M','F']

# Iterate through the rows and assign to the appropriate columns
for i in rows:
    if i in gender:
        count = 6
#     if '\n' in i:
#         count = 0
    if '\xa0' in i:
        count = 2
        
        
    if count == 0:
        no.append(i)
        count += 1
    elif count == 1:
        try:
            if isinstance(int(i), int):
                no.append(i)
                count = 1
        except:
            athlete.append(i)
            count += 1
    elif count == 2:
        nation.append(i)
        count += 1
    elif count == 3:
        sport.append(i)
        count += 1  
    elif count == 4:
        years.append(i)
        count += 1 
    elif count == 5:
        games.append(i)
        count += 1 
    elif count == 6:
        gender.append(i)
        count += 1 
    elif count == 7:
        gold.append(i)
        count += 1 
    elif count == 8:
        silver.append(i)
        count += 1 
    else:
        bronze.append(i)
        count = 0

data_df_base = {}
raw_data = [no, athlete, nation, sport, years, games, gender, gold, silver, bronze]

for index, value in enumerate(headers[:10]):
    data_df_base[value] = raw_data[index]

# Removed rows with inconsistent lengths
del data_df_base['No.']
del data_df_base['Athlete']
del data_df_base['Gender']

# 5 Create a table that shows for each country how many gold, silver, bronze, and total medals it won in that list.
# Convert to a dataframe
data_df = pd.DataFrame(data_df_base)

# Clean the dataframe
data_df['Bronze'] = data_df['Bronze'].str.replace('\n','') 
data_df.loc[:,['Gold', 'Silver', 'Bronze']] = data_df.loc[:,['Gold', 'Silver', 'Bronze']].astype(int)

# Create final output
data_df.groupby(['Nation'])[['Gold','Silver','Bronze']].agg('sum')
```
